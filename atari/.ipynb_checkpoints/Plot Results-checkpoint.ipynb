```python
%matplotlib inline
from matplotlib import pyplot as plt
import numpy as np
from scipy import stats
```

```python
NSpace = np.loadtxt("N-SpaceInvaders.csv", delimiter=',')
TSpace = np.loadtxt("T-SpaceInvaders.csv", delimiter=',')

NSpaceReward = NSpace[:,1]
NSpaceStep = NSpace[:,3]

TSpaceReward = TSpace[:,1]
TSpaceStep = TSpace[:,3]
```

```python
# Color wheel: https://www.sessions.edu/color-calculator/
# -----------------------------------------
fig = plt.figure(figsize=(5, 3))
plt.title("Space Invaders - Normal Train")

plt.plot(NSpaceStep, NSpaceReward, linestyle=':', color='#ff0000', 
         alpha=1, label='reward')

z1 = np.polyfit(NSpaceStep,NSpaceReward,1)
p1 = np.poly1d(z1)

plt.plot(NSpaceStep, p1(NSpaceStep), color='#22ff00', 
         linewidth=3.3, alpha=1, label="trend")

plt.ylabel("Reward")
plt.xlabel("Steps")
plt.legend(loc='best')
plt.xlim(0,1000000)
plt.show()
fig.savefig('space-n.pdf', bbox_inches='tight')
#-----------------------------------------
fig = plt.figure(figsize=(5, 3))
plt.title("Space Invaders - Transfer Learning")

plt.plot(TSpaceStep, TSpaceReward, linestyle=':', color='#0000ff', 
         alpha=1, label='reward')

z2 = np.polyfit(TSpaceStep,TSpaceReward,1)
p2 = np.poly1d(z2)

plt.plot(TSpaceStep, p2(TSpaceStep), color='#ff7d04', 
         linewidth=3.3, alpha=1, label="trend")

plt.ylabel("Reward")
plt.xlabel("Steps")
plt.legend(loc='best')
plt.xlim(0,1000000)
plt.show()
fig.savefig('space-t.pdf', bbox_inches='tight')
#-----------------------------------------
fig = plt.figure(figsize=(5, 3))
plt.title("Space Invaders - Trend Comparation")

plt.plot(range(0,1000000), p1(range(0,1000000)), color='red', 
         linewidth=3.3, alpha=1, label="normal trend")
plt.plot(range(0,1000000), p2(range(0,1000000)), color='blue', 
         linewidth=3.3, alpha=1, label="tranfer learning trend")

plt.ylabel("Reward")
plt.xlabel("Steps")
plt.legend(loc='best')
plt.xlim(0,1000000)
plt.show()
fig.savefig('space-c.pdf', bbox_inches='tight')
```

```python
space_random = 148.0
space_human = 1668.7

demon_random = 152.1
demon_human = 1971.0
```

```python
#Space Invaders
n_space1 = np.loadtxt("w_n_space.h5_SpaceInvaders-v0_0_episodes.csv", delimiter=',')
t_space1 = np.loadtxt("w_t_space.h5_SpaceInvaders-v0_0_episodes.csv", delimiter=',')
n_demon1 = np.loadtxt("w_n_demon.h5_SpaceInvaders-v0_0_episodes.csv", delimiter=',')
t_demon1 = np.loadtxt("w_t_demon.h5_SpaceInvaders-v0_0_episodes.csv", delimiter=',')

n_space2 = np.loadtxt("w_n_space.h5_SpaceInvaders-v0_1_episodes.csv", delimiter=',')
t_space2 = np.loadtxt("w_t_space.h5_SpaceInvaders-v0_1_episodes.csv", delimiter=',')
n_demon2 = np.loadtxt("w_n_demon.h5_SpaceInvaders-v0_1_episodes.csv", delimiter=',')
t_demon2 = np.loadtxt("w_t_demon.h5_SpaceInvaders-v0_1_episodes.csv", delimiter=',')

n_space_reward = np.append(n_space1[1:,1], n_space2[1:,1])
t_space_reward = np.append(t_space1[1:,1], t_space2[1:,1])
n_demon_reward = np.append(n_demon1[1:,1], n_demon2[1:,1])
t_demon_reward = np.append(t_demon1[1:,1], t_demon2[1:,1])

perf_space_n_space_reward = (np.mean(n_space_reward) - space_random)*100/space_human
perf_space_t_space_reward = (np.mean(t_space_reward) - space_random)*100/space_human
perf_space_n_demon_reward = (np.mean(n_demon_reward) - space_random)*100/space_human
perf_space_t_demon_reward = (np.mean(t_demon_reward) - space_random)*100/space_human

print ('Normal Space     :', 
       'Mean:', np.mean(n_space_reward), 
       #'Mode:', stats.mode(n_space_reward)[0][0],
       'Performance', (np.mean(n_space_reward) - space_random)*100/space_human, 
       'Std:', np.std(n_space_reward),
      )
print ('Transfer to Space:', 
       'Mean:', np.mean(t_space_reward), 
       #'Mode:', stats.mode(t_space_reward)[0][0],
       'Performance', (np.mean(t_space_reward) - space_random)*100/space_human,
       'Std:', np.std(t_space_reward),
      )
print ('Normal Demon     :', 
       'Mean:', np.mean(n_demon_reward),         
       #'Mode:', stats.mode(n_demon_reward)[0][0],
       'Performance', (np.mean(n_demon_reward) - space_random)*100/space_human,
       'Std:', np.std(n_demon_reward),
      )
print ('Transfer to Demon:', 
       'Mean:', np.mean(t_demon_reward),        
       #'Mode:', stats.mode(t_demon_reward)[0][0],
       'Performance', (np.mean(t_demon_reward) - space_random)*100/space_human,
       'Std:', np.std(t_demon_reward),
      )
```

```python
#Space Invaders
n_space1 = np.loadtxt("w_n_space.h5_DemonAttack-v0_0_episodes.csv", delimiter=',')
t_space1 = np.loadtxt("w_t_space.h5_DemonAttack-v0_0_episodes.csv", delimiter=',')
n_demon1 = np.loadtxt("w_n_demon.h5_DemonAttack-v0_0_episodes.csv", delimiter=',')
t_demon1 = np.loadtxt("w_t_demon.h5_DemonAttack-v0_0_episodes.csv", delimiter=',')

n_space2 = np.loadtxt("w_n_space.h5_DemonAttack-v0_1_episodes.csv", delimiter=',')
t_space2 = np.loadtxt("w_t_space.h5_DemonAttack-v0_1_episodes.csv", delimiter=',')
n_demon2 = np.loadtxt("w_n_demon.h5_DemonAttack-v0_1_episodes.csv", delimiter=',')
t_demon2 = np.loadtxt("w_t_demon.h5_DemonAttack-v0_1_episodes.csv", delimiter=',')

n_space_reward = np.append(n_space1[1:,1], n_space2[1:,1])

t_space_reward = np.append(t_space1[1:,1], t_space2[1:,1])

n_demon_reward = np.append(n_demon1[1:,1], n_demon2[1:,1])

t_demon_reward = np.append(t_demon1[1:,1], t_demon2[1:,1])


perf_demon_n_space_reward = (np.mean(n_space_reward) - demon_random)*100/demon_human
perf_demon_t_space_reward = (np.mean(t_space_reward) - demon_random)*100/demon_human
perf_demon_n_demon_reward = (np.mean(n_demon_reward) - demon_random)*100/demon_human
perf_demon_t_demon_reward = (np.mean(t_demon_reward) - demon_random)*100/demon_human
performance_demon_attack = [perf_demon_n_space_reward, perf_demon_t_space_reward, 
                            perf_demon_n_demon_reward, perf_demon_t_demon_reward]

print ('Normal Space     :', 
       'Mean:', np.mean(n_space_reward), 
       #'Mode:', stats.mode(n_space_reward)[0][0],
       'Performance', perf_demon_n_space_reward, 
       'Std:', np.std(n_space_reward),
      )
print ('Transfer to Space:', 
       'Mean:', np.mean(t_space_reward), 
       #'Mode:', stats.mode(t_space_reward)[0][0],
       'Performance', perf_demon_t_space_reward,
       'Std:', np.std(t_space_reward),
      )
print ('Normal Demon     :', 
       'Mean:', np.mean(n_demon_reward),         
       #'Mode:', stats.mode(n_demon_reward)[0][0],
       'Performance', perf_demon_n_demon_reward,
       'Std:', np.std(n_demon_reward),
      )
print ('Transfer to Demon:', 
       'Mean:', np.mean(t_demon_reward),        
       #'Mode:', stats.mode(t_demon_reward)[0][0],
       'Performance', perf_demon_t_demon_reward,
       'Std:', np.std(t_demon_reward),
      )
```
